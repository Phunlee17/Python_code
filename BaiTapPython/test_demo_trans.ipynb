```python
"""
Bai tap: 
Xay dung tu dien vơi nhung chưc nang sau
1. Thêm tư vưng
2. Tra cuu ý nghia
3. Cập nhật ý nghĩa của từ vững
4. xóa
5. Xóa all
6. end
"""
```

```python
#khai bao tu dien
tuDien = {}

while(True):
    print("Vui long lua chon")
    print(
    """
    1. Them tu vung va y nghia cua tu vung vao tu dien \n
    2. Tra cuu y nghia cua tu dien \n
    3. Cap nhat y nghia cho tu vung \n
    4. Xoa mot tu vung trong tu dien \n
    5. Xoa het tu vung trong tu dien \n
    6. Cho phep nguoi dung in ra tu vung \n
    7. Cho phep nguoi dung in ra tu vung duoi dang "Tu Vung" : "Y nghia" \n
    8. Ket thuc chuong trinh
    
    """)
    luaChon = int(input("Nhap su lua chon cua ban: "))
    
    if(luaChon == 1 or luaChon ==3):
        tuVung = input("Nhap tu vung: ")
        yNghia = input("Nhap y nghia cua tu vung: ")
        tuDien[tuVung] = yNghia
        print("Cap nhat du lieu thanh cong")
        
    elif(luaChon == 2):
        tuVung = input("Nhap tu ban muon tra cuu: ")
        print("y nghia cua tu ban muon tim:", tuDien[tuVung])
        
    elif(luaChon == 4):
        tuVung = input("Tu ban muon xoa: ")
        tuDien.pop(tuVung)
        print("Ban da xoa thanh cong")
        
    elif(luaChon == 5):
        tuDien.clear()
        print("Ban da xoa toan bo du lieu.")
        
    elif(luaChon == 6):
        print("Nhap tu ban muon in ra")
        for x in tuDien.keys():
            print(x)
            
    elif(luaChon == 7):
        print("Danh sach cac tu co trong tu dien")
        for x, y in tuDien.items():
            print(x, " : ", y)
            
    elif(luaChon == 8):
        break;
    else:
        print("Nhap su lua chon khong dung")
```
